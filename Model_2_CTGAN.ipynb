```python
#Loading the data
import pandas as pd
df = pd.read_csv('soil.csv')
df.head()
```

Label-Encoding the District column to make it numerical


```python

df.columns = df.columns.str.strip() # cutting of the trailing whitespaces to avoid confusion
print(df.columns.tolist())
```

```python
#Convert Text to Numbers using labelEncoder
from sklearn.preprocessing import LabelEncoder

lr = LabelEncoder()

df['District_id'] = lr.fit_transform(df['District'])
df = df.drop('District', axis=1)
```

```python
print(df)
```

```python
# Import sdv and check available single table GAN models
import sdv
display(dir(sdv.single_table))
```

```python
# Import the CTGAN synthesizer model (Conditional GAN for tabular data)
from sdv.single_table import CTGANSynthesizer
```

```python
# Import metadata generator
from sdv.metadata import SingleTableMetadata

# Create a metadata object
metadata = SingleTableMetadata()

# Automatically detect metadata (column types, formats, etc.)
metadata.detect_from_dataframe(df)

python_dict = metadata.to_dict()
```

```python
metadata.validate_data(data=df)
```

```python
from sdv.single_table import CTGANSynthesizer

synthesizer = CTGANSynthesizer(
    metadata,
    enforce_rounding=False,
    epochs=3000,
    verbose=True
)
```

```python
synthesizer.fit(df)
```

```python
synthetic_data = synthesizer.sample(num_rows=10000)
```

```python
synthetic_data
```

```python
import plotly.subplots as sp
import plotly.graph_objects as go
```

```python
from sdv.evaluation.single_table import get_column_plot

figs = []
for i,j in enumerate(list(df.columns)[:-1]) :

    fig = get_column_plot(
        real_data=df,
        synthetic_data=synthetic_data,
        column_name=j,
        metadata=metadata
    )
    figs.append(fig)

fig = sp.make_subplots(rows=2, cols=3, subplot_titles=list(df.columns)[:-1])

rows =2
columns = 3
for i in range(rows) :
    for j in range(columns) :
        count = i+j

        fig.add_trace(figs[count]['data'][0], row=i+1, col=j+1)
        fig.add_trace(figs[count]['data'][1], row=i+1, col=j+1)


fig.update_layout(height=600, width=900, title_text='2x3 Subplots')

fig.show()
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(
    df,
    synthetic_data,
    metadata)
```
